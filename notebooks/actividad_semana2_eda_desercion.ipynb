## 5. Relación entre variables y problema de deserción

En esta sección deberás reflexionar sobre qué variables podrían estar relacionadas con la deserción estudiantil.


```python
# Frecuencia de la variable objetivo
df['estado_final'].value_counts()
```

```python
# Promedio de asistencia según estado final
df.groupby('estado_final')['asistencia'].mean()
```

```python
# Promedio de notas según estado final
df.groupby('estado_final')['promedio_notas'].mean()
```

```python
# Distribución por tipo de ingreso y estado final
pd.crosstab(df['tipo_ingreso'], df['estado_final'])
```

## 6. Visualización básica

En esta sección realizarás algunas visualizaciones simples para apoyar la interpretación de los datos.


```python
# Histograma de asistencia
df['asistencia'].plot(kind='hist', bins=10)
plt.title('Distribución de asistencia')
plt.xlabel('Asistencia')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico de barras del estado final
df['estado_final'].value_counts().plot(kind='bar')
plt.title('Distribución de estado final')
plt.xlabel('Estado final')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Boxplot de promedio_notas por estado_final
df.boxplot(column='promedio_notas', by='estado_final')
plt.title('Promedio de notas según estado final')
plt.suptitle('')
plt.xlabel('Estado final')
plt.ylabel('Promedio notas')
plt.show()
```
