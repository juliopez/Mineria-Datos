## 4. Tipos de variables
Antes de comenzar el análisis, es importante distinguir dos grandes tipos de variables:

### Variables numéricas

Son variables que representan cantidades o medidas.

Ejemplos:

* edad
* ingresos
* total_compras
* visitas_web

Estas variables permiten calcular:

* promedio
* mediana
* desviación estándar
* valores mínimos y máximos

### describe ()

Una de las funciones más utilizadas en EDA es describe().

Esta función genera automáticamente estadísticas descriptivas para las variables numéricas del dataset.

Entre los principales resultados se encuentran:

* count: cantidad de registros válidos
* mean: promedio
* std: desviación estándar
* min: valor mínimo
* max: valor máximo
* 25%, 50%, 75%: percentiles

Estas medidas ayudan a comprender el comportamiento general de los datos.

```python
# Estadísticas descriptivas de variables numéricas
df.describe()
```

Por ejemplo:

* Un promedio muy alto podría indicar presencia de valores extremos.
* Una desviación estándar elevada sugiere alta dispersión.
* Diferencias importantes entre el valor máximo y el promedio pueden indicar outliers.

### Variables categóricas

Representan grupos o categorías.

Ejemplos:

* género
* región
* canal_preferido
* tipo_cliente

Estas variables normalmente se analizan mediante:

* frecuencia de aparición
* porcentajes
* gráficos de barras

```python
# Frecuencia de categorías
df["canal_preferido"].value_counts()
```

```python
# Frecuencia porcentual
df["canal_preferido"].value_counts(normalize=True) * 100
```

Este análisis permite detectar:

* categorías dominantes
* categorías poco frecuentes
* posibles inconsistencias en escritura

## 5. Tipos de gráficos

#### Histograma

La distribución de una variable muestra cómo se reparten los datos.

Analizar distribuciones permite:

* detectar concentraciones de datos
* identificar posibles valores atípicos
* observar asimetrías
* comprender patrones generales

Uno de los gráficos más utilizados para esto es el histograma.

```python
plt.hist(df["edad"], bins=10)
plt.title("Distribución de edad")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

En este histograma:

* el eje horizontal representa los valores de la variable;
* el eje vertical representa la frecuencia;
* las barras muestran cuántos registros existen en cada rango.

Si observamos barras muy alejadas del resto, podrían existir valores atípicos.

### Gráfico de barras

Los gráficos de barras permiten comparar categorías.

En este caso, podemos observar qué canal es utilizado con mayor frecuencia por los clientes.

```python

conteo = df["canal_preferido"].value_counts()

conteo.plot(kind="bar")
plt.title("Canal preferido")
plt.xlabel("Canal")
plt.ylabel("Cantidad")
plt.show()
```

### Dispersión

Además de analizar variables individualmente, el EDA busca detectar relaciones iniciales entre variables.

Por ejemplo:

* ¿Los clientes con mayores ingresos realizan más compras?
* ¿Existe relación entre edad y frecuencia de compra?

Estas relaciones pueden explorarse mediante gráficos de dispersión.

```python
plt.scatter(df["frecuencia_compra"], df["monto_total"])
plt.title("Ingresos vs Total de compras")
plt.xlabel("Ingresos")
plt.ylabel("Total compras")
plt.show()
```

Este tipo de gráfico permite observar:

* tendencias generales,
* agrupamientos,
* posibles relaciones positivas o negativas.
